```python
import pickle
scx=pickle.load(open("Standard_Scaler_X.pkl",'rb'))
scy=pickle.load(open("Standard_Scaler_Y.pkl",'rb'))
scy
```

```python
model=pickle.load(open("final_model.sav",'rb'))
model
```

```python
# user input 
age=int(input("Age:"))
bmi=float(input("BMI:"))
children=int(input("Children:"))
sex_male=int(input("Sex_Male: 0/1"))
smoker_yes=int(input("Smoker_yes 0 / 1: "))

# Transform the user input data using the input scaler
preInput=scx.transform([[age,bmi,children,sex_male,smoker_yes]])
preInput
```

```python
# Predict the standardized output using the trained model
result= model.predict(preInput)
result
```

```python
# Inverse transform the output to get the raw output prediction
preOutput=scy.inverse_transform([result])
print("Future Prediction : {}".format(preOutput))
```
